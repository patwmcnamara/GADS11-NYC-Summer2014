describe what the following methods of the array, `a`, do:

* shape - describes the shape of the matrix, the number of rows and columns
* dtype - provides the datatype that a is set for, in this case 64-bit signed integer
* flatten - converts a multi-dimensional matrix into a one-dimensional matrix, ordered by ascending row
* max/min - return the largest / smallest numeric value from the matrix
* argmax/argmin -  provide the relative memory location in that *flattened* matrix where the largest/smallest value is located
* argsort - provide a full list of relative memory location in each row where each value smallest to largest is provided.
* T - return the matrix turned so column becomes row and row becomes column a[3,2] becomes a[2,3]
* copy - generate and return a copy of the array for assignment
* diagonal - return the diagonal from a cell in the matrix to the opposite side of the matrix, providing each intermediate value

```python
b = np.array(a)
%timeit b.max()  # numpy array method
```
