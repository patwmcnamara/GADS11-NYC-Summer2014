```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Import standard data set
coffee = pd.read_csv('data.csv')
# Import data set with additional fields of review count, and average review score
coffee_reviews = pd.read_csv('data_with_review_count.csv')
```

```python
coffee.describe()
```

```python
coffee_reviews.head()
# Notice that Google Places will not assign an official review score until the number of reviews is >= 5
```

```python
# Which are the best coffee stores on official Google Places score (top 10)?
coffee_nonulls = coffee[pd.notnull(coffee['rating'])]
coffee_nonulls.sort(columns = ['rating'], ascending = False)[['rating', 'name', 'vicinity']].head(10)
```

```python
# Which are the best coffee stores on average reviewer score(top 10)?
coffee_reviews.sort(columns = ['rating_score'], ascending = False)[['rating_score','rating_count', 'name', 'vicinity']].head(30)
# Are places cheating high scores? We can check this if rating_count <5
```

```python
# Get all official scores where result is nonnull
all_scores = coffee_nonulls.rating
```

```python
# Generate first figure with plot of all scores in histogram
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)
all_scores.hist(ax=ax1,color='g', alpha=0.5, bins=30,range=(1,5))
plt.xlabel('Scores', fontsize=20)
plt.ylabel('Count', fontsize= 18)
plt.xlim((1,5))
plt.show()
```

```python
# Generate second figure with 5 histograms comparing average user reviews as a function of review number to the global stats
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(15, 8), tight_layout=True)
font = {'size'   : 14}
width = 0.5
for ax, num in zip(axes.ravel(), range(1,6)):
    # Make a new plot for each review count
    tmp = coffee_reviews[coffee_reviews['rating_count'] == num].rating_score
    b = np.arange(1,5+1.0/2.0,1.0/2.0)  
    ax.hist(coffee_nonulls['rating'], bins = b, histtype='stepfilled', normed=1, color='#cccccc', alpha=1)
    tmp.hist(ax = ax,bins = b, histtype='stepfilled', normed=1, color='r', alpha=0.4)
    ax.set_ylim([0, 1])
    ax.set_xlabel('Score',fontsize=14)
    title_str = '%d reviewer(s)' %(num)
    if num == 5:
           title_str = '%d+ reviewer(s)' %(num)
    ax.set_title(title_str,fontsize=18)
    ax.xaxis.set_ticklabels(['1', '', '2', '', '3', '', '4', '', '5'])
    # Change label font sizes
    for label in ax.xaxis.get_ticklabels():
        label.set_fontsize(14) 
        
    for label in ax.yaxis.get_ticklabels():
        label.set_fontsize(14) 
plt.show()
```
